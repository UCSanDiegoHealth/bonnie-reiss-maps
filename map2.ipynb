```python
import folium
import json
import pandas as pd
from branca.colormap import LinearColormap
import geopandas as gpd
from shapely.geometry import shape
import re

# Load the GeoJSON files
with open('ca_california_zip_codes_geo.min.json', 'r') as f:
    geojson_data = json.load(f)

with open('San_Diego_County_Boundary.geojson', 'r') as f:
    san_diego_geojson = json.load(f)

# Ensure missing values are filled in the dataframes
patient_df['%'] = patient_df['%'].fillna(0)
census_df['Asthma Pctl'] = census_df['Asthma Pctl'].fillna(0)

# Create dictionaries for data
zip_employees = dict(zip(patient_df['ZIP'].astype(str), patient_df['%']))
zip_asthma = dict(zip(census_df['ZIP'].astype(str), census_df['Asthma Pctl']))

zip_employees = {str(k): v for k, v in zip_employees.items()}
zip_asthma = {str(k): v for k, v in zip_asthma.items()}

# Convert San Diego GeoJSON to a GeoDataFrame
san_diego_gdf = gpd.GeoDataFrame.from_features(san_diego_geojson['features'])

# Convert California ZIP code GeoJSON to a GeoDataFrame
zip_gdf = gpd.GeoDataFrame.from_features(geojson_data['features'])

# Filter ZIP codes that intersect with San Diego County boundary
san_diego_boundary = san_diego_gdf.unary_union
zip_gdf['geometry'] = zip_gdf['geometry'].apply(shape)
san_diego_zip_gdf = zip_gdf[zip_gdf['geometry'].within(san_diego_boundary)]

# Extract San Diego ZIP codes
san_diego_zips = set(san_diego_zip_gdf['ZCTA5CE10'].astype(str))

# Filter ZIP codes present in both datasets and within San Diego County
filtered_zip_codes = san_diego_zips.intersection(zip_employees.keys()).intersection(zip_asthma.keys())

# Ensure that only ZIP codes with positive employee values are considered
valid_zip_employees = {zip_code: zip_employees[zip_code] for zip_code in filtered_zip_codes if zip_employees[zip_code] > 0}

# Recalculate valid asthma data for filtered ZIP codes
valid_zip_asthma = {zip_code: zip_asthma[zip_code] for zip_code in valid_zip_employees.keys()}

# Calculate weighted priority
weighted_priority = {zip_code: valid_zip_employees[zip_code] * valid_zip_asthma[zip_code] for zip_code in valid_zip_employees.keys()}

# Sort and get top 5 ZIP codes based on weighted priority
sorted_weighted_priority = sorted(weighted_priority.items(), key=lambda x: x[1], reverse=True)
top_5_priority_zip_codes = {zip_code for zip_code, _ in sorted_weighted_priority[:5]}

# Create a base map centered on San Diego County
m = folium.Map(location=[32.7157, -117.1611], zoom_start=10)

# Define a color scale with white for zero employees
color_scale = LinearColormap(['white', '#007DBA'], vmin=0, vmax=max(valid_zip_employees.values()))

# Add the choropleth layer for ZIP code shading (patient density) directly to the map (always visible)
folium.GeoJson(
    geojson_data,
    style_function=lambda feature: {
        'fillColor': color_scale(valid_zip_employees.get(feature['properties']['ZCTA5CE10'], 0)) if feature['properties']['ZCTA5CE10'] in valid_zip_employees else 'white',
        'color': 'black',
        'weight': 0.5,
        'fillOpacity': 0.7
    },
    tooltip=folium.GeoJsonTooltip(fields=['ZCTA5CE10'], aliases=['ZIP Code:'])
).add_to(m)

# Add San Diego County boundary directly to the map (always visible)
folium.GeoJson(
    san_diego_geojson,
    style_function=lambda x: {
        'color': 'black',
        'weight': 3,
        'fillColor': 'none',
        'fillOpacity': 0
    },
    tooltip='San Diego County'
).add_to(m)

# Function to calculate radius based on asthma severity
def get_asthma_radius(asthma_value):
    if asthma_value <= 0:
        return 0
    max_radius = 30
    asthma_normalized = asthma_value / max(zip_asthma.values()) if max(zip_asthma.values()) > 0 else 0
    return asthma_normalized * max_radius

# Add circle markers for asthma severity directly to the map (always visible)
for feature in geojson_data['features']:
    zip_code = feature['properties']['ZCTA5CE10']
    if zip_code in valid_zip_employees:
        asthma_value = zip_asthma.get(zip_code, 0)
        asthma_value_rounded = round(asthma_value, 2)
        
        coordinates = feature['geometry']['coordinates']
        if isinstance(coordinates[0][0], list):
            all_coords = [coord for polygon in coordinates for coord in polygon]
        else:
            all_coords = coordinates[0]
        
        avg_lat = sum([coord[1] for coord in all_coords if len(coord) == 2]) / len(all_coords)
        avg_lon = sum([coord[0] for coord in all_coords if len(coord) == 2]) / len(all_coords)

        # Check if this ZIP code is in the top 5 weighted priority areas
        is_top_5 = zip_code in top_5_priority_zip_codes
        marker_color = 'red' if is_top_5 else '#FFB500'

        folium.CircleMarker(
            location=[avg_lat, avg_lon],
            radius=get_asthma_radius(asthma_value),
            color=marker_color,
            fill=True,
            fill_color=marker_color,
            fill_opacity=0.6,
            tooltip=f'ZIP Code: {zip_code}<br>Asthma Percentile: {asthma_value_rounded}%{"<br><b>High Priority Area</b>" if is_top_5 else ""}'
        ).add_to(m)

# Extract geocodes from addresses (in a real scenario, you'd use geocoding)
# For this example, we'll use hardcoded coordinates for UCSD hospitals
hospital_coordinates = {
    'Hillcrest Medical Center': [32.7553, -117.1653],  # Coordinates for UCSD Hillcrest
    'Jacobs Medical Center': [32.8773, -117.2250],     # Coordinates for UCSD Jacobs
    'East Campus Medical Center': [32.7798, -117.0478]  # Coordinates for UCSD East Campus
}


clinic_coordinates = {
    'Garden View Road Clinic': [33.0689, -117.2507],
    'Encinitas Express Care': [33.0474, -117.2618],
    '4S Ranch Clinic': [33.0243, -117.0834],
    'Rancho Bernardo Express Care': [33.0247, -117.0840],
    'Downtown Express Care': [32.7158, -117.1626],
    'Bankers Hill Clinic': [32.7266, -117.1625],
    'Eastlake/Chula Vista Express Care': [32.6381, -116.9776],
    '4th & Lewis Medical Offices': [32.7503, -117.1621],
    'Hillcrest Psychiatry': [32.7560, -117.1610],
    'Kearny Mesa Obstetrics and Gynecology': [32.8159, -117.1534],
    'Moores Cancer Center': [32.8772, -117.2334],
    'PET/CT Center (Radiation Oncology)': [32.8780, -117.2321],
    'Shiley Eye Care – Hillcrest ': [32.7489, -117.1608],
    'Murrieta Clinic': [33.5629, -117.1886],
    'Medical Offices South': [32.7513, -117.1634],
    'UTC Clinic': [32.8709, -117.2173],
    'Chancellor Park - 4510 Executive Dr': [32.8741, -117.2144],
    'Chancellor Park - 4520 Executive Dr': [32.8744, -117.2143],
    'El Camino Real Clinic': [33.0582, -117.2588],
    'Sorrento Valley Clinic': [32.8820, -117.2107],
    'Pacific Highlands Ranch/Carmel Valley Express Care': [32.9595, -117.2370],
    'Encinitas Maternal Fetal Care': [33.0679, -117.2535],
    'Pediatrics': [32.7989, -117.1584],
    'University Center Lane Clinic': [32.8696, -117.2108],
    'La Jolla Urgent Care': [32.8698, -117.2456],
    'Surgical Services': [32.8702, -117.2118],
    'La Jolla Psychiatry': [32.8704, -117.2454],
    'Vista Express Care': [33.2085, -117.2386],
    'Genesee Ave Clinics': [32.8761, -117.2227],
    'Perlman Medical Offices': [32.8781, -117.2229],
    'Koman Family Outpatient Pavilion': [32.8787, -117.2237],
    'Shiley Eye Institutee': [32.8791, -117.2241],
    'Sulpizio Cardiovascular Center': [32.8785, -117.2210],
    'East Campus Office Building (ECOB)': [32.8786, -117.2207],
    'Altman Clinical and Translational Research Institute (ACTRI)': [32.8788, -117.2204],
    'Chula Vista Radiation Oncology': [32.6370, -116.9621],
    'Kearny Mesa Specialty Pharmacy': [32.8351, -117.1301],
    'Scripps Ranch Clinic': [32.9144, -117.1263]
}

# Function to extract ZIP code from address
def extract_zip_code(address):
    # Look for a 5-digit ZIP code in the address
    match = re.search(r'(\d{5})', address)
    if match:
        return match.group(1)
    return None

# Function to estimate coordinates for clinics based on ZIP code centroids
def get_coordinates_from_zip(zip_code):
    # Find the feature with this ZIP code
    for feature in geojson_data['features']:
        if feature['properties']['ZCTA5CE10'] == zip_code:
            coordinates = feature['geometry']['coordinates']
            if isinstance(coordinates[0][0], list):
                all_coords = [coord for polygon in coordinates for coord in polygon]
            else:
                all_coords = coordinates[0]
            
            avg_lat = sum([coord[1] for coord in all_coords if len(coord) == 2]) / len(all_coords)
            avg_lon = sum([coord[0] for coord in all_coords if len(coord) == 2]) / len(all_coords)
            
            return [avg_lat, avg_lon]
    return None

# Create hospital markers list
hospital_feature_group = folium.FeatureGroup(name="Hospitals")
hospital_label_feature_group = folium.FeatureGroup(name="Hospital Labels")

# Create hospital markers with hospital icon and tooltips, no shadow
for _, hospital in hospital_df.iterrows():
    hospital_name = hospital['Name']
    coordinates = hospital_coordinates.get(hospital_name)
    
    if coordinates:
        # Create a marker with hospital icon, no shadow
        marker = folium.Marker(
            location=coordinates,
            tooltip=folium.Tooltip(hospital_name, permanent=False),
            icon=folium.Icon(icon="h-square", prefix="fa", color="blue", shadowSize=[0, 0], class_name="awesome-marker hospital-icon")
        )
        marker.add_to(hospital_feature_group)
        
        # Create a text label for the hospital
        label = folium.Marker(
            location=coordinates,
            icon=folium.DivIcon(
                icon_size=(150, 36),
                icon_anchor=(75, 18),
                html=f'<div class="hospital-label" style="font-size: 10pt; font-weight: bold; color: blue;">{hospital_name}</div>'
            )
        )
        label.add_to(hospital_label_feature_group)

# Create clinic markers list
clinic_feature_group = folium.FeatureGroup(name="Clinics")

for _, clinic in clinic_df.iterrows():
    clinic_name = clinic['Name']
    coordinates = clinic_coordinates.get(clinic_name)
    
    if coordinates:
        marker = folium.Marker(
            location=coordinates,
            tooltip=folium.Tooltip(f"{clinic_name}", permanent=False),
            icon=folium.Icon(icon="plus", prefix="fa", color="green", shadowSize=[0, 0], class_name="awesome-marker clinic-icon")
        )
        marker.add_to(clinic_feature_group)

# Add feature groups to the map
hospital_feature_group.add_to(m)
hospital_label_feature_group.add_to(m)
clinic_feature_group.add_to(m)

# Custom JS for toggling markers and updating legend
toggle_js = """
<script>
    function toggleHospitals() {
        const checkbox = document.getElementById('hospital-toggle');
        const display = checkbox.checked ? 'block' : 'none';

        // Toggle hospital icon markers
        document.querySelectorAll('.hospital-icon').forEach(marker => {
            marker.style.display = display;
        });

        // Toggle hospital text labels
        document.querySelectorAll('.leaflet-marker-icon .hospital-label').forEach(label => {
            const parent = label.closest('.leaflet-marker-icon');
            if (parent) parent.style.display = display;
        });

        // Toggle legend
        const legend = document.getElementById('hospital-legend');
        if (legend) legend.style.display = display;
    }

    function toggleClinics() {
        const checkbox = document.getElementById('clinic-toggle');
        const display = checkbox.checked ? 'block' : 'none';

        // Toggle clinic icon markers
        document.querySelectorAll('.clinic-icon').forEach(marker => {
            marker.style.display = display;
        });

        // Toggle legend
        const legend = document.getElementById('clinic-legend');
        if (legend) legend.style.display = display;
    }

    document.addEventListener('DOMContentLoaded', function () {
        document.getElementById('hospital-toggle').checked = true;
        document.getElementById('clinic-toggle').checked = true;

        setTimeout(function () {
            toggleHospitals();
            toggleClinics();
        }, 500);
    });
</script>
"""

# Create a custom control panel for toggling hospitals and clinics
custom_control_html = """
<div id="custom-layer-control" style="
    position: fixed;
    top: 80px;
    left: 10px;
    background: white;
    padding: 10px;
    border: 2px solid rgba(0,0,0,0.2);
    border-radius: 4px;
    z-index: 1000;
    font-family: Arial, sans-serif;
    font-size: 12px;
    box-shadow: 0 1px 5px rgba(0,0,0,0.4);
">
    <div style="margin-bottom: 5px;"><b>Toggle Layers</b></div>
    <div>
        <input type="checkbox" id="hospital-toggle" onchange="toggleHospitals()" checked>
        <label for="hospital-toggle">Hospitals</label>
    </div>
    <div>
        <input type="checkbox" id="clinic-toggle" onchange="toggleClinics()" checked>
        <label for="clinic-toggle">Clinics</label>
    </div>
</div>
"""

# Add the custom control and toggle JavaScript to the map
m.get_root().html.add_child(folium.Element(custom_control_html))
m.get_root().html.add_child(folium.Element(toggle_js))

# Create HTML for hospital and clinic icons in the legend
hospital_icon_html = '<i class="fa fa-h-square fa-2x" style="color:blue;"></i>'
clinic_icon_html = '<i class="fa fa-plus fa-2x" style="color:green;"></i>'

# Modified Legend HTML with IDs for toggling hospital and clinic entries
legend_html = f'''
<div style="position: fixed; top: 70px; right: 50px; width: 350px; height: 350px; background-color: white; 
z-index:9999; font-size:14px; border:2px solid black; padding: 15px;">
    <p><b>Legend</b></p>
    <p><span style="display: inline-block; width: 12px; height: 12px; background-color: #FFB500; border-radius: 50%;"></span>: Asthma percentile relative to all California zip codes</p>
    <p><span style="display: inline-block; width: 12px; height: 12px; background-color: red; border-radius: 50%;"></span>: Top 5 weighted priority areas (patient % * asthma percentile)</p>
    <p id="hospital-legend" style="display: block;"><span style="display: inline-block; vertical-align: middle;">{hospital_icon_html}</span>: UC San Diego Health Hospitals</p>
    <p id="clinic-legend" style="display: block;"><span style="display: inline-block; vertical-align: middle;">{clinic_icon_html}</span>: UC San Diego Health Clinics</p>
    <p>Note: Larger dots indicate areas with higher rates of asthma</p>
    <p><a href="https://oehha.ca.gov/calenviroscreen/maps-data" target="_blank" style="color:blue;">Data From CalEnviroScreen 4.0</a></p>
</div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

# Adjusted Title HTML
title_html = '''
<div style="position: fixed; top: 10px; left: 50%; transform: translateX(-50%); 
background-color: white; z-index: 9999; padding: 10px; border: 2px solid black;
font-size: 18px; font-weight: bold; text-align: center; width: 400px;">
    Asthma & UC San Diego Health Patient Distribution: Identifying High Priority Areas
</div>
'''
m.get_root().html.add_child(folium.Element(title_html))

# Adjusted Employee Scale Legend HTML
employee_scale_legend_html = '''
<div style="position: fixed; top: 430px; right: 50px; width: 250px; height: 180px; background-color: white;
z-index:9999; font-size:14px; border:2px solid black; padding: 10px;">
    <p><b>Patient Distribution Legend</b></p>
    <p>Patient Density (as of 2023)</p>
    <div style="width: 100%; height: 10px; background: linear-gradient(to right, white, #00629B);"></div>
    <div style="display: flex; justify-content: space-between; margin-top: 5px;">
        <span>0%</span>
        <span>{:.2f}%</span>
    </div>
</div>
'''.format(round(max(valid_zip_employees.values()), 2))
m.get_root().html.add_child(folium.Element(employee_scale_legend_html))

# Save the map to an HTML file
m.save('san_diego_patient_asthma_map_with_hospitals_and_clinics.html')
```
